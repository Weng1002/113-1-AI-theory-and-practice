```python
import math

def entropy(p1,n1):
  if p1==0 and n1==0:
    return 1
  elif p1==0 or n1==0:
    return 0
  else:
    pp = p1/(p1+n1)
    np = 1 - pp
    return -pp * math.log2(pp) - np * math.log2(np)

def information_gain(p1,n1,p2,n2):
  nA = p1+n1
  nB = p2+n2
  n = nA+nB
  return entropy(p1+p2,n1+n2) - ((nA/n)*entropy(p1,n1) + (nB/n)*entropy(p2,n2))

print(information_gain(21,5,8,30))
print(information_gain(18,33,11,2))

```

```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
feature1 = iris.data[50:80][:]
feature2 = iris.data[100:130][:]
feature = np.concatenate((feature1,feature2),axis=0)
label = np.concatenate((np.zeros(30),np.ones(30)),axis=0)
```

```python
def ID3Train(feature,label):
  node = dict()
  node['data'] = range(len(label))
  tree = []
  tree.append(node)
  t = 0
  while t < len(tree):
    idx = tree[t]['data']
    if sum(label[idx])==0:
      tree[t]['leaf'] = 1
      tree[t]['decision'] = 0
    elif sum(label[idx])==len(idx):
      tree[t]['leaf'] = 1
      tree[t]['decision'] = 1
    else:
      bestIG = 0
      for i in range(feature.shape[1]):
        pool = list(set(feature[idx,i]))
        pool.sort()
        for j in range(len(pool)-1):
          thres = (pool[j] + pool[j+1])/2
          G1 = []
          G2 = []
          for k in idx:
            if feature[k,i] < thres:
              G1.append(k)
            else:
              G2.append(k)
          p1 = sum(label[G1]==1)
          n1 = sum(label[G1]==0)
          p2 = sum(label[G2]==1)
          n2 = sum(label[G2]==0)
          thisIG = information_gain(p1,n1,p2,n2)
          if thisIG > bestIG:
            bestIG = thisIG
            bestG1 = G1
            bestG2 = G2
            bestthres = thres
            bestf = i
      if bestIG>0:
        tree[t]['leaf'] = 0
        tree[t]['selectf'] = bestf
        tree[t]['threshold'] = bestthres
        tree[t]['child'] = [len(tree),len(tree)+1]
        node = dict()
        node['data'] = bestG1
        tree.append(node)
        node = dict()
        node['data'] = bestG2
        tree.append(node)
      else:
        tree[t]['leaf'] = 1
        tree[t]['decision'] = sum(label[idx])>len(idx)/2
    t += 1
  return tree

DT = ID3Train(feature,label)
print(DT)
```

```python
def ID3Test(DT,feature1):
  t = 0
  while DT[t]['leaf']==0:
    if feature1[DT[t]['selectf']] < DT[t]['threshold']:
      t = DT[t]['child'][0]
    else:
      t = DT[t]['child'][1]
  return DT[t]['decision']

for i in range(len(label)):
  print(ID3Test(DT,feature[i,:]))
```
