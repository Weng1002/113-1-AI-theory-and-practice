Deep NN

```python
import numpy as np
import keras
from keras import layers
```

```python
num_classes = 10
input_shape_dnn = (784,)
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
x_train_dnn = x_train.reshape((60000,784))
x_test_dnn = x_test.reshape((10000,784))
y_train = keras.utils.to_categorical(y_train,num_classes)
print(y_train.shape)
y_test = keras.utils.to_categorical(y_test,num_classes)
print(y_test.shape)
```

```python
model_dnn = keras.Sequential(
    [
        keras.Input(shape=input_shape_dnn),
        layers.Dense(8,activation='relu'),
        layers.Dense(4,activation='relu'),
        layers.Dense(4,activation='relu'),
        layers.Dense(4,activation='relu'),
        layers.Dense(10,activation='softmax')
    ]
)
model_dnn.summary()
```

```python
batch_size = 128
epochs = 10
model_dnn.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model_dnn.fit(x_train_dnn,y_train,batch_size=batch_size,epochs=epochs,validation_split=0.1)
```

```python
score = model_dnn.evaluate(x_test_dnn,y_test,verbose=0)
print('Test loss:',score[0])
print('Test accuracy:',score[1])
```

CNN

```python
input_shape_cnn = (28,28,1)
x_train_cnn = np.expand_dims(x_train,-1)
x_test_cnn = np.expand_dims(x_test,-1)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model_cnn = keras.Sequential(
    [
        keras.Input(shape=input_shape_cnn),

        layers.Conv2D(2, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(8, kernel_size=(3, 3), activation='relu'),
        layers.MaxPooling2D(pool_size=(2, 2)),

        layers.Dropout(0.3),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),

        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ]
)

model_cnn.summary()
```

```python
model_cnn.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model_cnn.fit(x_train_cnn,y_train,batch_size=batch_size,epochs=epochs,validation_split=0.1)
```

```python
score = model_cnn.evaluate(x_test_cnn,y_test,verbose=0)
print('Test loss:',score[0])
print('Test accuracy:',score[1])
```

Shallow model


```python
model_shallowdnn = keras.Sequential(
    [
        keras.Input(shape=input_shape_dnn),
        layers.Dense(8,activation='relu'),
        layers.Dense(10,activation='softmax')
    ]
)
model_shallowdnn.summary()
```

```python
model_shallowdnn.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model_shallowdnn.fit(x_train_dnn,y_train,batch_size=batch_size,epochs=epochs,validation_split=0.1)
```

```python
score = model_shallowdnn.evaluate(x_test_dnn,y_test,verbose=0)
print('Test loss:',score[0])
print('Test accuracy:',score[1])
```
