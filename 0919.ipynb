```python
def OptionProfit(pos,ST):
  callpos = pos[0]
  putpos = pos[1]
  profit = np.zeros(ST.shape)
  for position in callpos:
    K = position[0] #履約價
    C = CALLprice[K] #保證金
    N = position[1] #口數
    profit += (np.maximum(ST-K,0)-C)*N
  for position in putpos:
    K = position[0]
    P = PUTprice[K]
    N = position[1]
    profit += (np.maximum(K-ST,0)-P)*N
  return profit
ST = np.arange(20000,24000)
pos1=[[],[[21800,1],[22000,-2],[22200,1]]]
profit1 = OptionProfit(pos1,ST)
plt.plot(ST,profit1,'r-',[ST[0],ST[-1]],[0,0],'k--')
```
