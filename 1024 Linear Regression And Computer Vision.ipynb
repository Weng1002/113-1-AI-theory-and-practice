```python
import numpy as np
def F1(t):
  return 0.063*(t**3) - 5.284*(t**2) + 4.887*t + np.random.normal(0,1)
```

```python
N = 1000
A = np.zeros((N,5))
b = np.zeros((N,1))
for n in range(N):
  t = np.random.random()*100
  b[n] = F1(t)
  A[n,0] = 1
  A[n,1] = t
  A[n,2] = t**2
  A[n,3] = t**3
  A[n,4] = t**4
x = np.linalg.lstsq(A,b)[0]
print(x)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
BG = Image.open('/content/AI.jpg')
FG = Image.open('/content/Yun.jpg')
BG_data = np.asarray(BG).copy()
FG_data = np.asarray(FG)

[HB,WB,t] = BG_data.shape
[HF,WF,t] = FG_data.shape

x1 = 490
y1 = 148
x2 = 490
y2 = 294
x3 = 536
y3 = 141
x4 = 536
y4 = 301

p1 = 0
q1 = 0
p2 = 0
q2 = HF-1
p3 = WF-1
q3 = 0
p4 = WF-1
q4 = HF-1

A = np.zeros((8, 8))
b = np.zeros((8, 1))

A[0,:] = [-x1, -y1, -1, 0, 0, 0, x1*p1, y1*p1]
b[0,0] = -p1
A[1,:] = [0, 0, 0, -x1, -y1, -1, x1*q1, y1*q1]
b[1,0] = -q1
A[2,:] = [-x2, -y2, -1, 0, 0, 0, x2*p2, y2*p2]
b[2,0] = -p2
A[3,:] = [0, 0, 0, -x2, -y2, -1, x2*q2, y2*q2]
b[3,0] = -q2
A[4,:] = [-x3, -y3, -1, 0, 0, 0, x3*p3, y3*p3]
b[4,0] = -p3
A[5,:] = [0, 0, 0, -x3, -y3, -1, x3*q3, y3*q3]
b[5,0] = -q3
A[6,:] = [-x4, -y4, -1, 0, 0, 0, x4*p4, y4*p4]
b[6,0] = -p4
A[7,:] = [0, 0, 0, -x4, -y4, -1, x4*q4, y4*q4]
b[7,0] = -q4

x = np.linalg.lstsq(A, b)[0]
H = np.zeros((3, 3))
H[0,0] = x[0]
H[0,1] = x[1]
H[0,2] = x[2]
H[1,0] = x[3]
H[1,1] = x[4]
H[1,2] = x[5]
H[2,0] = x[6]
H[2,1] = x[7]
H[2,2] = 1

xy1 = np.ones((3,1))
pq1 = np.zeros((3,1))
for h in range(HF):
  for w in range(WF):
    xy1[0,0] = w
    xy1[1,0] = h
    pq1 = np.dot(H, xy1)

    p = int(pq1[0]/pq1[2])
    q = int(pq1[1]/pq1[2])

    if (p >= 0 and p < WF) and (q >= 0 and q < HF):
      BG_data[h, w, :] = FG_data[q, p, :]

plt.imshow(BG_data)
plt.axis('off')
plt.show()
```
